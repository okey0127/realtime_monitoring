```python
import time
import logging
import datetime
import os

# set path
#now_dir = os.path.dirname(os.path.abspath(__file__))
day = time.strftime('%Y-%m-%d', time.localtime(time.time()))

def delete_old_data():
  save_range = 28 # 4주
  #delete old date-server.log file
  for f in os.listdir('./log'):
    f = os.path.join('./log', f)
    if os.path.isfile(f):
      timestamp_now = datetime.datetime.now().timestamp()
      is_old = os.stat(f).st_mtime < timestamp_now-(save_range * 24 * 60 * 60)
      if is_old:
        try:
          os.remove(f)
          print(f, 'is deleted')
        except OSError:
          pass #오류 무시

  #server.log
  #detect old file, information and delete
  f_save = []
  with open(f'log/server.log', 'r') as fp:
      f1 = fp.readline() # read first line of server.log
      if f1 == '': #빈 파일이면 동작하지 않는다.
        return 0 # end function
      else:
        while True:
          if 'root' in f1: 
            break
          else:
            f1 = fp.readline() # read next line of server.log
            if f1 == '': # f1에서 더 이상 읽을것이 없으면 빈 문자열을 출력한다.
              return 0 # end function

        words = f1.split()
        old = words[0].split('-')
        old_date = datetime.date(int(old[0]), int(old[1]), int(old[2]))
        now_date = datetime.date(datetime.datetime.today().year, datetime.datetime.today().month, datetime.datetime.today().day) 
        diff = int(str(now_date - old_date).split()[0]) # save difference of date
        fl = fp.readlines() # read all lines of server.log
        for line in fl:
          if 'root' in line:
            words = line.split()
            d = words[0].split('-')
            if len(d) == 3:
              date = datetime.date(int(d[0]),int(d[1]),int(d[2]))
            else:
              continue
            f_diff = int(str(now_date - date).split()[0])
            if f_diff <= save_range: #4주 내의 데이터만 저장한다.
              f_save.append(line)
  with open(f'log/server.log', 'w') as fp:
    fp.writelines(f_save)

delete_old_data()
```
